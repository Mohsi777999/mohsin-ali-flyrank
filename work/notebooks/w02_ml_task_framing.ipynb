## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
## My lane as an ML task

# I chose the Content Refresh / Content Opportunity Scoring lane.
# This is primarily a ranking (scoring) task because the goal is to prioritize webpages
# based on how likely they are to need a content refresh. Instead of simply predicting yes or no,
# the system assigns a score so content editors can review the highest-priority pages first.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
## Target or proxy

# The target is whether a page is showing signs of decline. In the starter dataset,
# this is represented by the is_declining_label, which is derived from the observed
# trend information. This serves as a proxy target for learning which pages should be
# prioritized for review. During modeling, the label is predicted, while the columns
# used to create it are not used as input features to avoid data leakage.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
## Success metric

# The main success metric is Precision@50, which measures how many of the top 50 pages
# recommended by the model actually belong to the target group. A higher Precision@50 means
# the content team spends more time reviewing pages that truly deserve attention instead
# of wasting effort on lower-priority pages.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
## Why ML beats a fixed rule here

# A simple rule, such as refreshing every page with low CTR or poor average position,
# cannot capture the full picture because page performance depends on many factors working together.
# Machine learning can combine multiple signals such as CTR, impressions, engagement, average position,
# content age, and other features to produce better priority scores. The goal is to support content
# teams with data-driven recommendations rather than relying on one fixed rule.
```
